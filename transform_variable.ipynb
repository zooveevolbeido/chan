```python

import math
import numpy as np
e = np.e
pi = np.pi

r = 0.5

def func(x):

    return 1/r * x**(-1/r-1)

##############################################


def transform_equ(X, y):

    Y = 1/r*np.log(X)      - y
    # don't forget (-y)
    return Y


y = 0.01
print("y(inp) = ", y)

#################################


ini_x = 1
n_simu = 15000
lr = 0.01
amp = 1

def sol(equ, lr, ini_x, amp ,n_simu, y):

    fir = transform_equ(ini_x, y)
    sec = transform_equ(ini_x + lr, y)
    val_con = [fir, sec]
    ans = []
    xs = [ini_x  ,ini_x + lr]


    for i in range(n_simu):

        cur_x = xs[-1] + lr#*amp
        cur_val = transform_equ(cur_x, y)

        two_ago = val_con[-2] #equ( xs[-2] )
        one_ago = val_con[-1] #equ( xs[-1] )

        if abs(one_ago) < abs(two_ago)  and  abs(one_ago) < abs(cur_val):
            ans.append(xs[-1])
            xs.append(cur_x)
            val_con.append(cur_val)

        else:
            xs.append(cur_x)
            val_con.append(transform_equ(cur_x, y))


    return ans#, val_con

x = sol(transform_equ, lr, ini_x , amp ,n_simu, y)[0]
print("x(inverse) = ",x)

######################################


def derivative(t, s):

    return ( transform_equ(t+s, y) - transform_equ(t, y) ) / s

s = 0.00001
dy_over_dx = derivative(x, s)
print('dy_over_dx = ', dy_over_dx)
print()

dx_over_dy = 1/ dy_over_dx
print("dx_over_dy = ",dx_over_dy)
print()

############################################

#mu = 0
#std = 1

gy = func(x)

print("gy = ",gy)
print()


fy = gy * dx_over_dy
print("fy = ",fy)
print()

#print(np.log(3))
#zz = 1/(np.sqrt(2*pi)*std ) * e**(-1 *  ((x-mu)**2/2*std**2)     )

#print(zz)

###################################


print()
print()


```
